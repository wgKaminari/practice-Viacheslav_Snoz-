```python
import requests
import re
```

```python
#В мене працює тіки з англ вікіпедією ://
url = "https://en.wikipedia.org/wiki/Green_Book_(film)"
text_from_page = requests.get(url)

all_text = text_from_page.text
all_text
```

```python
release_date = re.findall(r"Release date?.*plainlist.*>\n.*(\d{4})", all_text)
release_date
```

```python
director = re.findall(r"Directed by.*?<a.*?>(.*?)</a>", all_text)
director
```

```python
genre_pattern = re.compile(r'\b(?:action|drama|science\sfiction|horror)\b', re.IGNORECASE)
matches = genre_pattern.findall(all_text)
matches[:1]
```

```python
pattern = re.compile(r'\b([0-9,.]+)\s*million\b', re.IGNORECASE)
matches = pattern.findall(all_text)
matches
```
